```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

years = np.arange(2024, 2031)

growth_rates = {
    "AI / ML": 0.25,
    "Cloud Engineering": 0.15,
    "Cybersecurity": 0.18,
    "Data Science / Analytics": 0.12,
    "DevOps / SRE": 0.10,
    "Software Development (general)": 0.08
}

start_index = 100
data = {}
for role, r in growth_rates.items():
    vals = [start_index * ((1 + r) ** (y - 2024)) for y in years]
    data[role] = vals

df = pd.DataFrame(data, index=years)
df.index.name = "Year"

# --- Graph ---
plt.figure(figsize=(10,6))
for col in df.columns:
    plt.plot(df.index, df[col], marker='o', label=col)
plt.title("Projected Relative Demand for IT Skills in India (2024â€“2030, Index=100)")
plt.xlabel("Year")
plt.ylabel("Demand Index (2024 = 100)")
plt.grid(True, linestyle='--', linewidth=0.5)
plt.legend(loc='upper left', fontsize='small')
plt.tight_layout()
plt.show()

# Optional: save to CSV
df.to_csv("it_skill_demand_index_2024_2030.csv", index=True)
```
